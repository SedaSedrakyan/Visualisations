## This project shows versions of interactive visualisations which can be implemented on different kinds of data, however considering confidentiality factor I am presenting all those visualisations on randomly initialized or already published data.

```python
import pandas as pd
import numpy as np
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode,plot, iplot
%matplotlib inline
```

```python
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns = 'A B C D'.split())
```

```python
df2 = pd.DataFrame({'Category': ['A','B','C'], 'Values': [32,43,50]})
```

```python
df.iplot(kind = 'scatter', x ='A', y='B', mode ='markers', size = 30)
```

```python
df2.iplot(kind ='bar',x = 'Category', y='Values')
```

```python
df.count().iplot(kind = 'bar')
```

```python
df.sum().iplot(kind = 'bar')
```

```python
df.iplot(kind = 'box')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5],'y':[10,20,30,20,10], 'z': [500,400,300,200,100]})
```

```python
df3.iplot(kind = 'surface')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5],'y':[10,20,30,20,10], 'z': [5,4,3,2,1]})
```

```python
df3.iplot(kind = 'surface',colorscale = 'rdylbu')
```

```python
df['A'].iplot(kind='hist',bins=50)
```

```python
df.iplot(kind='hist')
```

```python
df[['A','B']].iplot(kind ='spread')
```

```python
df.iplot(kind = 'bubble',x = 'A', y = 'B', size= 'C')
```
